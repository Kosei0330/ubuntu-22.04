```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure(figsize=(5,5), facecolor="w", edgecolor="none")
graphs = []
flag_legend = True   ### 凡例描画のフラグ
for i in range(201):
    x,u = np.loadtxt(f"./data/data{i:04d}.dat" , unpack=True)
    
    obj = []
    a, = plt.plot(x,u, linewidth=0.0, marker="o", markersize=1, color="black", label="numerical solution")
    if flag_legend:   ### 一回のみ凡例を描画
        plt.legend(loc="upper right")
        flag_legend = False
    obj += [a]

    graphs.append(obj)

### 1秒間に10枚出力させるような動画の作成
ani = animation.ArtistAnimation(fig, graphs, interval=10,blit=True)
### いったんgifで出力
ani.save("animation.gif", writer="pillow")

### 12.6秒で出力できた。
### 凡例を出そうとすると約1分ほど追加でかかる。そうでもないかも。
```
