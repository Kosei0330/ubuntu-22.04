```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

t,x,y = np.loadtxt("3_1.txt" , unpack=True)
print(t.shape, x.shape, y.shape)

plt.figure(figsize=(6,6))   ### 正方形に出力する
plt.xlim(-2,2)   ### 範囲の変更
plt.ylim(-2,2)
plt.plot(x,y, label="Runge-Kutta solution")

plt.legend(loc="upper right")   ### 凡例を右上に出す
```

```python
### アニメーションを見るためにはanimation.gif を参照。
### 下に表示される画像は一枚の画像になってしまっている。
### 動画の情報をすべて含んだ配列。graph[1]には1番目のフレームの情報がはいる。
graphs = []

### 画像化の共通情報。画像サイズ、背景色、枠色など。
fig = plt.figure(figsize=(5,5), facecolor="w", edgecolor="none")

### 厳密解（半径1の円）を出力するための配列
theta = np.linspace(0, 2*np.pi, 100)


### 各出力時刻のスナップショットの作成
for i in range(0,100):
    
    obj = []
    ### 変数の直後にカンマを入れているのは、
    ### plt.plot(…)がリストを返しているので、その一つ目の要素を取りますよ、という話
    ### 厳密解の出力
    im_exact_solution, = plt.plot(np.cos(theta), np.sin(theta), linewidth=1.0, color="black")
    obj += [im_exact_solution]

    ### 数値計算解の出力
    im, = plt.plot(x[i],y[i], linewidth=0.0, marker="o", markersize=4, color="red")
    obj += [im]

    graphs.append(obj)
    

### 1秒間に10枚出力させるような動画の作成
ani = animation.ArtistAnimation(fig, graphs, interval=10,blit=True)
### いったんgifで出力
ani.save("animation.gif", writer="pillow")

plt.show()
```
