```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure(figsize=(5,5), facecolor="w", edgecolor="none")
plt.xlim(-0.1,1.1)   ### 範囲の変更
plt.ylim(-1.1,1.1)
graphs = []
for i in range(201):
    x,u = np.loadtxt(f"./data/data{i:04d}.dat" , unpack=True)
    
    obj = []
    a, = plt.plot(x,u, linewidth=0.0, marker="o", markersize=1, color="black")
    obj += [a]

    graphs.append(obj)

### 1秒間に10枚出力させるような動画の作成
ani = animation.ArtistAnimation(fig, graphs, interval=10,blit=True)
### いったんgifで出力
ani.save("animation.gif", writer="pillow")

### 12.0秒で出力できた。
```
