```python
### python 3.10.13 を選択すること。
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ファイルからデータを読み込む
data = np.loadtxt('report_1a.txt')

# データの形状を確認する
nx, ny = data.shape

# メッシュグリッドを作成する
x = np.linspace(0, nx-1, nx)
y = np.linspace(0, ny-1, ny)
X, Y = np.meshgrid(x, y)

# 3次元プロットを作成する
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, data, cmap='viridis')

# ラベルを設定する
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Value')

# プロットを表示する
plt.show()
```
