```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

t,x,y = np.loadtxt("3_1.txt" , unpack=True)
print(t.shape, x.shape, y.shape)

plt.figure(figsize=(6,6))   ### 
plt.xlim(-2,2)   ### 
plt.ylim(-2,2)
plt.plot(x,y, label="Runge-Kutta solution")

plt.legend(loc="upper right")   ### 
```

```python
### animation.gif 
### 
### graph[1]1
graphs = []

### 
fig = plt.figure(figsize=(5,5), facecolor="w", edgecolor="none")

### 1
theta = np.linspace(0, 2*np.pi, 100)


### 
for i in range(0,100):
    
    obj = []
    ### 
    ### plt.plot()
    ### 
    im_exact_solution, = plt.plot(np.cos(theta), np.sin(theta), linewidth=1.0, color="black")
    obj += [im_exact_solution]

    ### 
    im, = plt.plot(x[i],y[i], linewidth=0.0, marker="o", markersize=4, color="red")
    obj += [im]

    graphs.append(obj)
    

### 110
ani = animation.ArtistAnimation(fig, graphs, interval=10,blit=True)
### gif
ani.save("animation.gif", writer="pillow")

plt.show()
```
